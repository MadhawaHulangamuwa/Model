```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Madhawa Hulangamuwa\Desktop\Model\FD_America.xlsx")
```

```python
df.head(40)
```

```python
df = df.drop(['ifs', 'code','country','imf_region','imf_income','FI','FM','FID','FIA','FIE','FMD','FMA','FME'], axis = 1)
```

```python
df.info()
```

```python
df.head()
```

```python
df.set_index('year', inplace=True)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn')

# line plot - the yearly average air temperature in Barcelona
df.plot(color='green', linewidth=3, figsize=(12,6))

# modify ticks size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend('')

# title and labels
plt.title('The FD_index of America', fontsize=20)
plt.xlabel('year', fontsize=16)
plt.ylabel('FD_index', fontsize=16)
```

```python
df['SMA_10'] = df.FD.rolling(10, min_periods=1).mean()
df['SMA_5'] = df.FD.rolling(5, min_periods=1).mean()
```

```python
# colors for the line plot
colors = ['green', 'red', 'purple']

# line plot -
df.plot(color=colors, linewidth=3, figsize=(12,6))

# modify ticks size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(labels =['FD_index','10-years SMA', '5-years SMA'], fontsize=14)

# title and labels
plt.title('Average FD_index', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('FD_index', fontsize=16)
```
